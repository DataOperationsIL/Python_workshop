```python
def divides_by_three(num):
    if num % 3 == 0:
        return True
    else:
        return False

#alternative:
def divides_by_three(num):
    return num % 3 == 0
```

```python
divides_by_three(-4)
```

```python
my_list = [1, 4, 6, 7, 9, 9, 9]

for x in my_list:
    if divides_by_three(x) and x%2==0:
        print(x)
```
