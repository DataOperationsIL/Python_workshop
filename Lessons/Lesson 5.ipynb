# Weekly problem

Guessing game from here: https://www.practicepython.org/exercise/2014/04/02/09-guessing-game-one.html

Generate a random number between 1 and 9 (including 1 and 9). Ask the user to guess the number, then tell them whether they guessed too low, too high, or exactly right. 

Extras:
Keep the game going until the user types “exit”
Keep track of how many guesses the user has taken, and when the game ends, print this out.

```python
import random 
```

```python
#input function
x = input("hi write something:")
```

```python
type(x)
```

```python
random_number = random.randint(1,9) # number that user needs to guess
user_guess = 0
num_guesses = 0 
```

```python
random_number
```

```python
#basic version
while user_guess != random_number: #while user didn't yet guess correctly
    user_guess = input("enter guess:")
    user_guess = int(user_guess) #turning string to number
    if user_guess > random_number:
        print ("too high")
    elif user_guess < random_number:
        print ("too low")
    else:
        print("correct")
    
```

```python
random_number = random.randint(1,9) # number that user needs to guess
user_guess = 0
num_guesses = 0 
```

```python
random_number
```

```python
# adding 'exit' and guess count

while user_guess != random_number and user_guess != 'exit': #while user didn't yet guess correctly AND user didn't write 'exit'
    user_guess = input("enter guess:")
    
    if user_guess == 'exit':
        print("you made " +str(num_guesses) + " guesses")
        break
    else:    
        user_guess = int(user_guess) #turning string to number
        if user_guess > random_number:
            num_guesses += 1 #other option: num_guesses = num_guesses + 1
            print ("too high")
        elif user_guess < random_number:
            num_guesses += 1
            print ("too low")
        else:
            print("correct")
            print("it took you "+ str(num_guesses) + " guesses")
```

## Grouping and Sorting

```python
df.groupby('parent_title')['rand_number'].sum()
```

```python
df.groupby('parent_title')['rand_number'].mean().reset_index()
```

```python
df.groupby('parent_title')['rand_number'].agg(['sum', 'min', 'mean']).reset_index()
```

```python
df.groupby('parent_title')['rand_number'].agg(['sum', 'min', 'mean']).reset_index().sort_values(by='parent_title', ascending=False)
```
